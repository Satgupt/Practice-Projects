### Aproach1

```python
from sklearn.preprocessing import LabelEncoder
l_enc = LabelEncoder()
```

```python
for i in data.columns:
    data[i] = l_enc.fit_transform(data[i])
data.head()
```

```python
data_corr = data.corr().abs() #to get the correlation values

plt.figure(figsize=(15,13))
sns.heatmap(data_corr,vmin=-1,vmax=1,annot=True,square=True,center=0,fmt='.1g',linewidth=0.02)

plt.tight_layout()
```

### Split features and label

```python
X = data.drop(columns = 'class')
Y = data['class']
```

### Normalization
  => using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
x_scaled = Scaler.fit_transform(X)
```

### Splitting: Test and Train

```python
#importing libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,Y,test_size = 0.22, random_state=53)
```

## Machine Learning
#### Model1 : Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier 

dt_c= DecisionTreeClassifier()
dt_c.fit(x_train, y_train) #training the data
```

```python
dt_c.score(x_train,y_train) #checking accuracy score of the trained data
```

```python
y1_pred_dec = dt_c.predict(x_test) #predicting the label value for the test feature using the model
```

### Evaluating the Decision Tree Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score,classification_report
```

```python
confusion_matrix(y_test,y1_pred_dec)
```

```python
pd.DataFrame(confusion_matrix(y_test,y1_pred_dec),columns=['Q3','Q4'],index=['Q3','Q4'])
```

#### Hyper Parameter Optimization
Using GridSearchCV

```python
grid_param = {'criterion':['gini','entropy'],
          'max_depth':range(2,10,2),
          'min_samples_leaf':range(2,10,2),
          'min_samples_split':range(2,10,2)}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=dt_c,param_grid=grid_param,cv=5,n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameters = grid_search.best_params_
best_parameters
```

```python
dt_c = DecisionTreeClassifier(criterion='gini',max_depth=10,min_samples_leaf= 2,min_samples_split= 2)
```

```python
dt_c.fit(x_train,y_train)
dt_c.score(x_test,y_test)
```

```python
y1_pred = dt_c.predict(x_test)
```

```python
accuracy_score(y_test,y1_pred)
```

```python
print(classification_report(y_test,y1_pred))
```

### Model3: LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train) #training the model
```

```python
y1_pred_log = log_reg.predict(x_test) #predicting the test data
```

```python
accuracy_score(y_test,y1_pred_log)
```

```python
fpr,tpr,thresholds = roc_curve(y_test,y1_pred_log)
```

```python
auc=roc_auc_score(y_test,y1_pred_log)
auc
```

```python
#plotting auc curve for logistic regression model
plt.plot(fpr,tpr,color='Orange',label='ROC')
plt.plot([0,1],[0,1],color='Blue',linestyle='--',label='ROC Curve (area= %0.3f)'%auc )
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operater Characteristics (ROC) Curve')
plt.legend()
plt.show()
```
